### Part 2 â€‘ Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active during the day. On weekends, there is reasonable activity in both cities.

However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend to be exclusive to each city. The Ultimate managers of city operations for the two cities have proposed an experiment to encourage driver partners to be available in both cities, by reimbursing all toll costs.

1) What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

2) Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:

    a) how you will implement the experiment

    b) what statistical test(s) you will conduct to verify the significance of the observation

    c) how you would interpret the results and provide recommendations to the city operations team along with any caveats.

Note: The two cities of Gotham and Metropolis are not in the provided dataset; however, you do
not need this information to answer Part 2.

#### Answer:

Since the purpose of this proposal is to encourage driver partner to not be exclusive to one city, I would choose the difference between pre-reimbursement and post-reimbursement number of times drivers have crossed the bridge in a week as the key measure of success of this experiment. The reason why I want to observe the data for a week is because the daily cycles in each city is different on weekdays and weekends. 

I'd observe and record the data for 10 weeks and conduct a one-sided t-test and assume the reimbursement doesn't increase the number of times drivers cross the bridge.  
**Null hypothesis**: The observed increase in the number of times drivers have cross the bridge in a week before reimbursement and after reimbursement is due to chance (and thus not due to reimbursement).  
**Alternate hypothesis**: The observed increase in the number of times drivers have cross the bridge in a week before reimbursement and after reimbursement is not due to chance (and is actually due to reimbursement).  
I'm also going to pick a significance level of 0.05.

Given this assumption, I should actually be able to get a difference in the number of times drivers have cross the bridge in a week before reimbursement and after reimbursement that's pretty similar to the one we actually got just by: a. shuffling the "cross bridge counts" column, b. keeping the reimbursement column the same, c. calculating the difference between the number of times drivers have cross the bridge in a week before reimbursement and after reimbursement.  
I would make the shuffle more useful by doing it many times, each time calculating the the difference between these number of times drivers crossing the bridge.  
I could then take the mean of all these differences, and this will be called our permutation difference. This permutation difference will be great indicator of what the difference would be if our initial assumption were true and reimbursement really doesnâ€™t impact on number of times drivers crossing the bridge.  
Then I would do a comparison. If the observed difference looks just like the permutation difference, then we stick with the claim that actually, reimburesment doesnâ€™t increase the number of time drivers crossing the bridge. If instead, however, the permutation difference differs significantly from the observed difference, we'll conclude: the toll reimbursement is infact effetive, and it increses the number of times drivers crossing the bridge.

#### 3.1 Data Cleaning and Exploratory Analysis

```python
#check for missing values
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by=['count','%'],ascending= False)
```

Goodnews, the overall data looks clean except some missing values in the columns 'avg_rating_of_driver','phone', and 'avg_rating_by_driver'.

Since our target is the retention of the users, I'll first identify which users were â€œactiveâ€ (i.e. took a trip) in the preceding 30 days.

```python
# Get latest date that data was pulled
df['last_trip_date'] = pd.to_datetime(df['last_trip_date'])
date_pulled = max(df['last_trip_date'])

# Get earliest active date to meet "retained" criteria
preceeding30 = date_pulled - dt.timedelta(days=30)
print(preceeding30)
```

```python
# create a column named 'retained': 1 for retained users and 0 for not retained users
df['retained'] = df['last_trip_date'].apply(lambda x: 1 if x>=preceeding30 else 0)

# get fraction of the observed users that were retained
fraction_retained = len(df[df['retained']==1])/len(df)
print('There were {}% of observed users retained.'.format(round(fraction_retained*100,2)))

# Create barplot for retention
df['retained'].astype('category').value_counts().plot(kind='bar')
plt.show()
```

The fraction of the observed users were retained is 37.61%. I'll then drop 'last_trip_date' and 'signup_date', since these were used to determine whether the user was retained and are no longer useful. 

```python
# drop 'last_trip_date' and 'signup_date'
df.drop(['signup_date','last_trip_date'],axis=1,inplace=True)
```

```python
# convert boolean values in 'ultimate_black_user' to 0 or 1
df.ultimate_black_user = df.ultimate_black_user.map({True:1,False:0})

# Display data distribution for all numerical features
df.hist(figsize=(15,10))
plt.subplots_adjust(hspace=0.5)
```

```python
# Retention in each city
sns.countplot(x='city',data=df,hue='retained')
plt.show()
```

The countplot above shows that King's Landing was the city that had the least users; however, there were more active users than inactive users in King's Landing. It's the opposite in the other two cities.

Among all observed users, there were more users using iphone than Android. Also, it seems that people using iPhone were more likely to retain.

#### 3.2 Pre-processing and Training

```python
# impute missing values in 'avg_rating_by_driver' and 'avg_rating_of_driver' using median
df['avg_rating_by_driver'].fillna(df['avg_rating_by_driver'].median(), inplace=True)
df['avg_rating_of_driver'].fillna(df['avg_rating_of_driver'].median(), inplace=True)
# impute missing values in 'phone' using the most frequent value
df['phone'].fillna(df['phone'].mode()[0], inplace=True)
```

```python
# check if there's any more missing values in our dataset
df.isna().any()
```

```python
# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['retained']), 
                                                    df.retained, test_size=0.3, random_state=47)
```

```python
df.dtypes
```

```python
#convert 'city','phone', and 'ultimate_black_user' to categorical data
df['city'] = df['city'].astype("category")
df['phone'] = df['phone'].astype("category")
df['ultimate_black_user'] = df['ultimate_black_user'].astype("category")

# make pipeline to pre-process data
categorical_features = ["city", "phone", "ultimate_black_user"]
numerical_features = ["avg_dist", "avg_rating_by_driver", "avg_rating_of_driver", "surge_pct", "trips_in_first_30_days", "weekday_pct"]

preprocessor = ColumnTransformer(
    transformers=[("num", Pipeline([("scaler", StandardScaler())]), numerical_features), 
                  ("cat", Pipeline([("encoder", OneHotEncoder())]), categorical_features)])

pipeline = Pipeline(steps=[("pre", preprocessor), 
                           ("rf", RandomForestClassifier(random_state = 37))])

#grid search parameters
max_depth = [int(x) for x in np.linspace(5, 15, num = 3)]
max_depth.append(None)
rf_param_grid = {
    'rf__max_depth': max_depth,
    'rf__n_estimators': [int(x) for x in np.linspace(start = 50, stop = 500, num = 5)]
}
```

#### 3.3 Modeling

```python
#fitting the pipeline to the grid search
rf_grid_cv = GridSearchCV(pipeline, param_grid=rf_param_grid, cv=5, n_jobs=-1)
rf_grid_cv.fit(X_train, y_train)
```

```python
rf_grid_cv.best_params_
```

```python
rf_estimator = rf_grid_cv.best_estimator_
```

```python
# Define scoring metrics
scoring = ['accuracy','precision', 'recall', 'f1']
# Use Cross-validation to evaluate the model
cv_results = cross_validate(rf_estimator, X_train, y_train, cv=5,scoring=scoring)
```

```python
# Print the mean cross validation score for each metrics
for key in cv_results:
    print(key,': ', np.mean(cv_results[key]))
```

```python
# Assess model performance on our test set
y_pred = rf_estimator.predict(X_test)
print('Model accuracy on test set: ',metrics.accuracy_score(y_test, y_pred))
print('Model f1 score on test set: ',metrics.f1_score(y_test, y_pred))
```

The score on test set is comparable to our training set score, which means that our model have not overfitted.

#### 3.4 Feature Importance

We extract feature importance to get a better understanding of what factors are the best predictors for retention.

```python
pipeline.fit(X_train, y_train)
importances = pipeline.named_steps['rf'].feature_importances_
print(importances)
```

```python
# getting feature names by replicating the transformers of our pipeline
num_cols = df[numerical_features].columns.tolist()
cat_cols = pd.get_dummies(df[categorical_features]).columns.tolist()
cols = num_cols + cat_cols
print(cols)
```

```python
#Plot a barplot of the random forest's feature importances (display the top 30 most important features only)
plt.subplots(figsize=(10, 5))
rf_feat_imps = pd.Series(importances, index=cols).sort_values(ascending=False)
rf_feat_imps.plot(kind='bar')
plt.xlabel('features')
plt.ylabel('importance')
plt.title('Best random forest classifier feature importances');
```

Most important features appear to be avg_dist, weekday_pct, avg_rating_by_driver, and 'surge_pct'.

In other words, to predict retention, the most important things to look at are the average distance in miles per trip, the percent of the userâ€™s trips occurring during a weekday,  the riderâ€™s average rating over all of their trips
, and the percent of trips taken with surge multiplier > 1.
